```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(-3, 3, 21)
y = np.linspace(-3, 3, 21)
X, Y = np.meshgrid(x, y)
dy = X - Y
dx = np.ones(dy.shape)

#color = dy
#lw = 0.1
plot1 = plt.streamplot(X,Y,dx, dy, color=dy, density=1., cmap='jet', arrowsize=1)
```

### Intro til numerisk derivasjon

```python


t = np.linspace(0, 5, 501)
p = -10*t**2+50*t
v = 50-2*10*t

plt.figure(figsize=(9, 4))
plt.subplot(1,2,1)
plt.plot(t, p, label="$y(t)$")
plt.plot([2], p[200], 'C0o', linewidth=0.0, markersize=6)
plt.plot([0, 5], [p[200] - v[200]*2, p[200] + v[200]*3], 'C3--', label="Tangent")
plt.xlabel("Tid $t$ (sekund)")
plt.ylabel("HÃ¸yde (meter)")
plt.legend(loc="lower center")
plt.ylim(ymax=80)
plt.title("Vertikal posisjon over tid $y(t)$")
yticks = list(plt.gca().get_yticks())
yticklabels = list(plt.gca().get_yticklabels())
yticklabels[0::2] = [""]*len(yticklabels[0::2])
plt.gca().set_yticks(yticks, yticklabels)
plt.subplot(1,2,2)
plt.plot(t, v, 'C3', label="$v(t)$")
plt.plot([2], v[200], 'C3o', linewidth=0.5, markersize=6)
plt.xlabel("Tid $t$ (sekund)")
plt.ylabel("Hastighet (m/s)")
plt.legend(loc="lower center")
yticks = list(plt.gca().get_yticks())
yticklabels = list(plt.gca().get_yticklabels())
yticklabels[0::2] = [""]*len(yticklabels[0::2])
yticks.append(v[200])
yticklabels.append(str(round(v[200])))
plt.gca().set_yticks(yticks, yticklabels)
plt.title("Hastighet over tid $v(t)$")
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 5, 501)
p = -10*t**2+50*t
v = 50-2*10*t

p_diff = (p[250]-p[200])/0.5

plt.figure(figsize=(9, 4))
plt.subplot(1,2,1)
plt.plot(t, p, label="$y(t)$")
plt.plot([2, 2.5], [p[200], p[250]], 'C2o', linewidth=0.0, markersize=6)
plt.plot([0, 5], [p[200] - v[200]*2, p[200] + v[200]*3], 'C3:', linewidth=0.7, label="Tangent")
plt.plot([0, 5], [p[200] - p_diff*2, p[200] + p_diff*3], 'C2--', label="Sekant")
plt.xlabel("Tid $t$ (sekund)")
plt.ylabel("HÃ¸yde (meter)")
plt.legend(loc="lower center")
plt.ylim([50, 70])
plt.xlim([0.9, 3.1])
yticks = list(plt.gca().get_yticks())
yticklabels = list(plt.gca().get_yticklabels())
yticklabels[1::2] = [""]*len(yticklabels[1::2])
plt.gca().set_yticks(yticks, yticklabels)
plt.title("Vertikal posisjon over tid $y(t)$")


plt.subplot(1,2,2)
plt.plot(t, v, 'C3', label="$v(t)$")
plt.plot([2], p_diff, 'C2o', linewidth=0.0, markersize=6)
plt.xlabel("Tid $t$ (sekund)")
plt.ylabel("Hastighet (m/s)")
plt.legend(loc="lower center")
plt.ylim([-20, 40])
plt.xlim([0.9, 3.1])
yticks = list(plt.gca().get_yticks())
yticklabels = list(plt.gca().get_yticklabels())
yticklabels[1::2] = [""]*len(yticklabels[1::2])
yticks.append(p_diff)
yticklabels.append(str(round(p_diff)))
plt.gca().set_yticks(yticks, yticklabels)
plt.gca().set_yticks(yticks, yticklabels)
plt.title("Hastighet over tid $v(t)$")
```

### Animasjon

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython import display

t = np.linspace(0, 5, 501)

p = lambda t : -10*t**2+50*t
v = lambda t : 50 - 2*10*t
h = 0.5
x = 2
p_diff = (p(2+h)-p(2))/h

fig = plt.figure(figsize=(9, 4))
ax1 = plt.subplot(1,2,1)
ax1.plot(t, p(t), label="$y(t)$")
diff_points, = ax1.plot([x, x+h], [p(x), p(x+h)], 'C2o', linewidth=0.0, markersize=6)
ax1.plot([0, 5], [p(x) - v(x)*x, p(x) + v(x)*(5-x)], 'C3:', linewidth=0.7, label="Tangent")
secant, = ax1.plot([0, 5], [p(x) - p_diff*x, p(x) + p_diff*(5-x)], 'C2--', label="Sekant")
ax1.set_xlabel("Tid $t$ (sekund)")
ax1.set_ylabel("HÃ¸yde (meter)")
ax1.legend(loc="lower center")
ax1.set_ylim([50, 70])
ax1.set_xlim([0.9, 3.1])
yticks = list(ax1.get_yticks())
yticklabels = list(ax1.get_yticklabels())
yticklabels[1::2] = [""]*len(yticklabels[1::2])
ax1.set_yticks(yticks, yticklabels)
ax1.set_title("Vertikal posisjon over tid $y(t)$")


ax2 = plt.subplot(1,2,2)
ax2.plot(t, v(t), 'C3', label="$v(t)$")
v_est, = ax2.plot([x], p_diff, 'C2o', linewidth=0.0, markersize=6)
ax2.set_xlabel("Tid $t$ (sekund)")
ax2.set_ylabel("Hastighet (m/s)")
ax2.legend(loc="lower center")
ax2.set_ylim([-10, 30])
ax2.set_xlim([0.9, 3.1])
yticks = list(ax2.get_yticks())
yticklabels = list(ax2.get_yticklabels())
yticklabels[1:-1] = [""]*len(yticklabels[1:-1])
yticks.append(p_diff)
yticklabels.append(str(round(p_diff,2)))
ax2.set_yticks(yticks, yticklabels)
ax2.set_title("Hastighet over tid $v(t)$")


def update(frame):
    global yticks
    global yticklabels
    h = 0.5*(10-frame)/10
    p_diff = (p(2+h)-p(2))/h
    diff_points.set_xdata([x, x+h])
    diff_points.set_ydata([p(x), p(x+h)])
    secant.set_ydata([p(x) - p_diff*x, p(x) + p_diff*(5-x)])
    v_est.set_ydata([p_diff])
    yticks[-1]=p_diff
    yticklabels[-1] = str(round(p_diff, 2))
    ax2.set_yticks(yticks, yticklabels)
    return diff_points, secant, v_est, ax2

# Create the animation
ani = FuncAnimation(fig, update, frames=range(0,10), blit=True)

# Display the animation
display.display(ani)
ani.save('differentiation.gif', writer='pillow', fps=1)
```
